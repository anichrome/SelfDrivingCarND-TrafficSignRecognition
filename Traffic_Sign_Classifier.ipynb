### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.

from sklearn.utils import shuffle
import tensorflow as tf
import cv2

def introduce_translation(image):
    rows, columns,_ = image.shape
    
    # define maximum translation in pixels
    max_translate = 2
    
    # choose random translation
    dx, dy = np.random.randint(-max_translate, max_translate, 2)

    # translation matrix
    M = np.float32([[1,0,dx], [0,1,dy]])
    
    # translate
    transformed_image = cv2.warpAffine(image, M, (columns,rows))
    transformed_image = transformed_image[:, :, np.newaxis]
    
    return transformed_image

def introduce_rotation(image):
    rows = image.shape[0]
    columns = image.shape[1]
    
    # define maximum rotation in degrees
    max_rotate_degree = 15
    
    # choose randomly rotation
    rotate_degree = np.random.randint(-max_rotate_degree, max_rotate_degree, 1)

    # roation matrix
    M = cv2.getRotationMatrix2D((columns / 2, rows / 2),rotate_degree, 1)
    
    # rotate
    transformed_image = cv2.warpAffine(image, M, (columns,rows))
    transformed_image = transformed_image[:, :, np.newaxis]
    
    return transformed_image

def introduce_scaling(image):
    rows, columns,_ = image.shape
    
    delta = np.random.randint(-2,2)

    # four transformed corners
    transformed_image_corners = np.float32(
        [[delta, delta], 
         [rows - delta, delta], 
         [delta, columns - delta], 
         [rows - delta, columns-delta]])

    # all four image corners 
    image_corners = np.float32([[0, 0], [rows, 0], [0, columns], [rows, columns]])

    # transformation matrix
    M = cv2.getPerspectiveTransform(transformed_image_corners,image_corners)

    # scale
    transformed_image = cv2.warpPerspective(image, M, (rows, columns))
    transformed_image = transformed_image[:, :, np.newaxis]
    
    return transformed_image

def add_gaussian_noise(image):

    shape = [image.shape[0], image.shape[1], image.shape[2]]
    x = tf.placeholder(dtype = tf.float32, shape = shape)# Adding Gaussian noise
    noise = tf.random_normal(shape=tf.shape(x), mean=0.0, stddev=1.0,
    dtype=tf.float32)
    output_image = tf.add(x, noise)
    
    return output_image

def local_normalize(images):
    # allocate space for proccessed images
    processed_images = np.ndarray((images.shape[0], images[0].shape[0], images[0].shape[1], 1), dtype=np.uint8)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(5,5))
    
    # process each image
    for i, image in enumerate(images):        
        image = clahe.apply(image)
        image = np.expand_dims(image, axis=2)
        processed_images[i] = image
    return processed_images

def global_normalize(images):
    mean = np.mean(images)
    std_var = np.std(images)
    normalized_images = np.copy(images)
    normalized_images = normalized_images.astype('float32')
    normalized_images -= mean
    
    if (std_var != 0.0):
        normalized_images /= std_var
    
    return normalized_images

def get_luma_channel(images):
    # allocate space for proccessed images
        
    processed_images = np.ndarray((images.shape[0], images[0].shape[0], images[0].shape[1], 1), dtype=np.uint8)
    
    # process each image
    for i, image in enumerate(images):
        # convert the image to YUV color space and get the luma chanel
        image = cv2.cvtColor(image, cv2.COLOR_RGB2YUV)
        image = cv2.split(image)[0]
        #image = cv2.split(cv2.cvtColor(image, cv2.COLOR_RGB2YUV))[0]
        image = np.expand_dims(image, axis=2)
        processed_images[i] = image
    return processed_images

def get_new_data(images, labels):
    samples_count = 5 

    # allocate space for proccessed images
    processed_images = np.ndarray((images.shape[0] * samples_count, images[0].shape[0], images[0].shape[1], 1), dtype=np.uint8)
    new_labels = np.ndarray((images.shape[0] * samples_count, 1), dtype=np.uint8)
    
    # process each image
    for i, image in enumerate(images):
        processed_images[i * samples_count] = image
        label = labels[i]
        new_labels[i * samples_count] = label 
        
        for j in range(1,samples_count):
            new_image = introduce_translation(image)
            new_image = introduce_rotation(new_image)
            new_image = introduce_scaling(new_image)
            #new_image = add_gaussian_noise(new_image)
            processed_images[i * samples_count + j] = new_image
            new_labels[i * samples_count + j] = label
            
    return processed_images, new_labels
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.

from tensorflow.contrib.layers import flatten

keep_prob = tf.placeholder(tf.float32)

def LeNet(x):    
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    features1 = 108
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, features1), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(features1))
    conv1   = tf.nn.bias_add(tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID'), conv1_b)

    # Activation.
    conv1 = tf.nn.relu(conv1)

    # Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # Layer 2: Convolutional. Output = 10x10x16.
    #features2 = 200
    features2 = 108
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, features1, features2), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(features2))
    conv2   = tf.nn.bias_add(tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID'),conv2_b)
    
    # Activation.
    conv2 = tf.nn.relu(conv2)

    # Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # dropout
    conv1 = tf.nn.dropout(conv1, keep_prob)
    conv2 = tf.nn.dropout(conv2, keep_prob)

    # Flatten. Input = 5x5x16. Output = 400.
    fc0   = tf.concat([flatten(conv2),flatten(conv1)], 1)
    flattenFeatureCount = 5 * 5 * features2 + 14 * 14 * features1
    
    # Layer 3: Fully Connected. Input = 400. Output = 84.
    features3 = 100 - n_classes
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(flattenFeatureCount,  features3), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(features3))
    fc2    = tf.nn.bias_add(tf.matmul(fc0, fc2_W), fc2_b)
    
    # Activation.
    fc2 = tf.nn.relu(fc2)
    fc2 = tf.nn.dropout(fc2, keep_prob)
    
    # Layer 4: Fully Connected. Input = 84. Output = Traffic Sign Classes
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(features3, n_classes), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(n_classes))
    logits = tf.nn.bias_add(tf.matmul(fc2, fc3_W),fc3_b)
            
    return logits
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.

x = tf.placeholder(tf.float32, (None, image_shape[0], image_shape[1], 1))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)

# Traning Pipeline
EPOCHS = 20
BATCH_SIZE = 32

rate = 0.0002

logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)


# Model Evaluation
# Evaluate how well the loss and accuracy of the model for a given dataset.

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()
validation_accuracy_figure = []

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
# Train the Model
# Run the training data through the training pipeline to train the model.
# Before each epoch, shuffle the training set.
# After each epoch, measure the loss and accuracy of the validation set.
# Save the model after training.

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    best_accuracy = 0.0
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.4})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        validation_accuracy_figure.append(validation_accuracy)
        
        if (validation_accuracy > best_accuracy):
            best_accuracy = validation_accuracy
            saver.save(sess, './lenet-traffic-signs-model')
            print("Model saved")
            print()
            
with tf.Session() as sess:
    print("Restore the best model from disk.")
    saver.restore(sess, "./lenet-traffic-signs-model")
    
    # Calculate Test Accuracy
    test_acc = evaluate(X_test, y_test)
    print('Testing Accuracy: {}'.format(test_acc))
    
plt.plot(validation_accuracy_figure)
plt.title("Validation Accuracy")
plt.show()
````
